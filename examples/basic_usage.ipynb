```python
import tissuemix as tm
import matplotlib.pyplot as plt
```

```python
# select number
a = 3
b = 3

# load target patches
target = tm.load_image(f'../images/targets/target_{str(a).zfill(2)}.png')
target_mask = tm.load_mask(f'../images/targets/target_{str(a).zfill(2)}_mask.png')

# load background patches
background = tm.load_image(f'../images/backgrounds/background_{str(b).zfill(2)}.png')
background_mask = tm.load_mask(f'../images/backgrounds/background_{str(b).zfill(2)}_mask.png')

# show target and background patches
tm.show_pair(target, target_mask, background, background_mask)
```

```python
x, _ = tm.apply_tissuemix(
    target=target,
    target_mask=target_mask,
    background=background,
    background_mask=background_mask,
    blend=False,
    color=False,
    warp=False)

plt.imshow(x)
plt.title('Basic mixing')
plt.show()

x, _ = tm.apply_tissuemix(
    target=target,
    target_mask=target_mask,
    background=background,
    background_mask=background_mask,
    blend=False,
    color=True,
    warp=False)

plt.imshow(x)
plt.title('Color correction')
plt.show()

x, _ = tm.apply_tissuemix(
    target=target,
    target_mask=target_mask,
    background=background,
    background_mask=background_mask,
    blend=True,
    color=True,
    warp=False)

plt.imshow(x)
plt.title('Blending')
plt.show()
```
